### Détection de Valeurs Aberrantes

```python
df_malaria.keys()
```

```python
df_malaria.info()
```

```python
df_malaria = df_malaria.select_dtypes(include=['number'])  # Garde uniquement les colonnes numériques
```

```python
for col in df_malaria.columns:
    if df_malaria[col].dtype in ['int64', 'float64']:  # Vérifie que la colonne est bien numérique
        q1 = df_malaria[col].quantile(0.25)
        q3 = df_malaria[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        median_value = df_malaria[col].median()
        
        df_malaria.loc[(df_malaria[col] < lower_bound) | (df_malaria[col] > upper_bound), col] = median_value
```

### Entraînement du Modèle de Régression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Créer le modèle de régression linéaire
model = LinearRegression()
```

```python
model
```

```python
# Entraîner le modèle
model.fit(X_train, y_train)
```

```python
# Prédictions sur les données de test
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Évaluer le modèle avec la MSE (Erreur Quadratique Moyenne) et le R²
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse}')
print(f'R² Score: {r2}')
```

```python
# Utiliser les meilleurs paramètres pour prédire et évaluer le modèle final
best_model = grid_search.best_estimator_
y_pred_best = best_model.predict(X_test)
```

```python
# Calculer le MSE et R² du meilleur modèle
mse_best = mean_squared_error(y_test, y_pred_best)
r2_best = r2_score(y_test, y_pred_best)

print(f'Mean Squared Error (MSE) du meilleur modèle: {mse_best}')
print(f'R² Score du meilleur modèle: {r2_best}')
```
