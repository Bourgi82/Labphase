```python
df_malaria = df_malaria.select_dtypes(include=['number'])  # Garde uniquement les colonnes numÃ©riques
```

```python
for col in df_malaria.columns:
    if df_malaria[col].dtype in ['int64', 'float64']:  # VÃ©rifie que la colonne est bien numÃ©rique
        q1 = df_malaria[col].quantile(0.25)
        q3 = df_malaria[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        median_value = df_malaria[col].median()
        
        df_malaria.loc[(df_malaria[col] < lower_bound) | (df_malaria[col] > upper_bound), col] = median_value
```
